# EXPLORATORY DATA ANALYSIS

```python
# Which manuacturer produces cereals with the highest calories?
# Group the manufacturers and calories to calculate the total distribution

df.groupby('Manufacturer')['calories'].sum()
```

```python
# Name the dataframe
df1= df.groupby('Manufacturer')['calories'].sum()
```

```python
plt.figure(figsize= (8,6))
df1.plot(kind='bar', x= 'mfr', ylabel= 'Total', title='OVERALL DISTRIBUTION OF CALORIES')
plt.show()  
```

Among the manufacturers listed, Kellogg's appears to produce cereals with the highest total calories, totaling 2500 calories.  There is notable variation in the total caloric contribution across manufacturers, ranging from as low as 100 calories for American Home Food Products to as high as 2500 calories for Kellogg's.  Manufacturers such as General Mills, Post, and Quaker Oats have relatively higher total caloric contributions compared to others like Nabisco and American Home Food Products. This suggests that they may cater to consumers seeking cereals with higher energy content or different nutritional profiles.

```python
# Nutritional content by each manufacturer
# Grouping by 'Manufacturer' and computing the average of each nutrient
average_nutritional_content = df.groupby('Manufacturer')[
    ['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins']].mean()

# Printing the average nutritional content by each manufacturer
print(average_nutritional_content)
```

```python
# Grouping by 'Manufacturer' and computing the mean of each nutrient
average_nutritional_content = df.groupby('Manufacturer')[
    ['calories', 'protein', 'fat', 'sodium', 'fiber', 'carbo', 'sugars', 'potass', 'vitamins']].mean().reset_index()

# Reshape the DataFrame for plotting
melted_df = average_nutritional_content.melt(id_vars='Manufacturer', var_name='Nutrient', value_name='Average')

# Plotting using Seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x='Nutrient', y='Average', hue='Manufacturer', data=melted_df)
plt.title("AVERAGE NUTRITIONAL CONTEBT BY MANUFACTURER")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

The average caloric content varies across manufacturers, with Kellogg's having the highest average calories (108.70), followed closely by Ralston Purina (115.00).  Nabisco stands out with the lowest average sugar content (1.83), indicating a potentially healthier option for consumers looking to reduce sugar intake. Nabisco also has the highest average fiber content (4.00), suggesting a focus on fiber-rich products. General Mills has the highest average sodium content (200.45), which may be a concern for individuals monitoring their sodium intake. Quaker Oats, on the other hand, has a relatively lower average sodium content (92.50).Kellogg's and Nabisco have higher average potassium content, indicating a potential focus on products with added potassium. American Home Food Products and Ralston Purina have the highest average vitamin content, suggesting an emphasis on fortified cereals.

```python
# Group by 'Manufacturer' and count occurrences of 'shelf' values
shelf_counts = df.groupby('Manufacturer')['shelf'].value_counts().unstack()

# Plot stacked bar chart using Seaborn
sns.set_style('whitegrid')  # Set style for better aesthetics
shelf_counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Number of Cereals on Each Shelf by Manufacturer')  
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Shelf')
plt.show()
```

Kelloggs is the manufacturer with the highest number of cereals and shelf 3 is the highest proportion amongst all the manufacturer except American Home Food Products
