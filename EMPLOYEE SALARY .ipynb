### VISUALIZATION

#### SALARY BY EXPERIENCE LEVEL

```python
# How does salary affect experience level?
df.groupby('experience_level')['salary'].mean().sort_values(ascending= False)

# Replace experience levels with full names
df['experience_level'].replace(
    {'EN': 'Entry Level', 'EX': 'Executive Level', 'MI': 'Mid Level', 'SE': 'Senior Level'}, inplace=True)
```

```python
# Let's name it Dataframe 2
df2 = df.groupby('experience_level')['salary'].mean()
df2
```

```python
# Plot a bar chart
df2.plot()
df2.plot(kind = 'bar')
plt.title('AVERAGE SALARY BY EXPERIENCE LEVEL')
```

Our analysis shows that Mid-level employees is the peak in average salary. Employees at mid-level positions command the highest average salaries of 480,617. The average salary for Entry-Level positions is approximately 264.622, reflecting the common trend where organizations offer lower initial salaries to employees with limited experience and skillset. The senior level experience category has a lower average salaries which could be because of the differences in job responsibilities or roles or even lack of opportunity for them to advance their skills.

### PROPORTION OF EMPLOYMENT TYPES FROM (2020-2022)

```python
# Group by 'work_year' and 'employment_type' and calculate the counts
employment_type_counts = df.groupby(['work_year', 'employment_type']).size()

# Group by 'work_year' and calculate the total count of employees for each year
total_employees_per_year = df.groupby('work_year').size()

# Calculate the proportion of each employment type within each year
employment_type_proportion = employment_type_counts.div(total_employees_per_year, level='work_year')

# Print the result
print(employment_type_proportion)
```

```python
# Assuming df10 is the dataframe
df10= employment_type_proportion
df10
```

```python
# Define colors for each employment type
colors = ['blue', 'green', 'red']

# Plot a bar chart for each year
df10.plot(kind = 'bar', color=colors)
plt.title('PROPORTION OF EMPLOYMENT TYPES FROM (2020-2022)')
```

Across all three years (2020-2022), full-time employment consistently represents the majority proportion of employment types.
In 2020, full-time employment accounts for approximately 94.4% of the workforce, followed by part-time employment (2.8%), contract (1.4%), and full employment (1.4%).
Similarly, in 2021 and 2022, full-time employment maintains its dominance, representing around 94.9% and 98.7% of the workforce, respectively.
